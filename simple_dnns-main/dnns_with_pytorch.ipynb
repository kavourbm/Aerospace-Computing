# Dataset description

## Available Classes, Methods, and Objects

#### Datasets/Classes: Univariate, Bivariate, SimpleData, NoiseData

#### Objects/Variables: layers, epochs, learning_rate, momentum, p_train
`layers: list`
- A list of integers; Eg.: [1, 2, 2, 1]
- 1st number is dependent on the dataset type; 1 for univariate; 2 for Bivariate etc...
- last number is always 1 in our datasets; we are only training regression
- middle numbers represent hidden layers

`epochs`: `int`
- Number of epochs/iterations to train the model
    
`learning_rate`: `float`
- Positive value between 0 and 1
    
`momentum`: `float`
- Use this to speed up training process; specifically targets gradient descent
- Value in the range of [0, 1)
    
`p_train`: `float`
- percent of data for training set
- Value in range of [0, 1]

#### Methods or Functions: set_parameters, train, restart, continue_train, validation; validation_real_data (only available in SimpleData)
`set_parameters`: returns `None`
- use to set parameters for the model training

`train`: returns `None`
- Trains the neural network model
    
`restart`: returns `None`
- Resets the model weights and biases to default parameters
    
`continue_train`: returns `None`
- Reads in a saved model and continues the training

`validation`: return `None`
- prints validation error for the model
   
`validation_real_data`: returns `None` (Only available in SimpleData)
- plots the model on real data to compare; runs validation method as well

# Univariate Dataset

```python
from univariate import Univariate
```

```python
model1 = Univariate()
```

```python
model1.data_set.plot()
```

```python
model1 = Univariate()

model1.data_set.plot()

model1.epochs = 2
model1.learning_rate = 0.07
model1.layers = [1, 4, 4, 4, 1]
model1.momentum = 0
model1.p_train = 0.7

model1.set_parameters()

model1.train(save_at = 100)  # Generates residual files at every 100 iterations
```

```python
model1.validation()
```

```python
model1.epochs = 201
model1.continue_train()

model1.train()
```

```python
model1.epochs = 1501
model1.learning_rate = 0.05
model1.set_parameters()
model1.continue_train()

model1.train()
```

```python
model1.epochs = 1501
model1.learning_rate = 0.01
model1.set_parameters()
model1.continue_train()

model1.train()
```

```python
model1.validation()  # Generally for these datasets the error should be close to zero
```

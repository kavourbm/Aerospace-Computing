```python
# Imports of Fun

# These two are to create the directory and filenames of the charts
from pathlib import Path

# This is to specify the location of my modules
import sys as syeet
syeet.path.append(str(Path().cwd())+"/"+"Modules")

# Numpy and PyPlot
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# My library
import kavouras1 as k
import textbook as t

# Inline outputs of matplotlib
%matplotlib inline

#
x,y = sp.symbols('x y')
```

```python
# Problem 1
f1 = 1/sp.sqrt(sp.sin(x))
F1 = sp.Integral(f1,(x,0,sp.pi/4))
print(F1.evalf())
```

```python
# Problem 2
f2 = ((x**4)*(sp.E**x))/((sp.E**x)-1)**2
F2 = np.array([])
for i in range(0,20):
    u = i/20
    if u == 0:
        F2 = np.append(F2,0)
    else:
        temp = sp.Integral((u**3)*f2,(x,0,1/u))
        F2 = np.append(F2,temp.evalf())
plt.plot(np.arange(0,1,.05),F2)
plt.show()
```

```python
# Problem 3
f3 = sp.ln(x)/((x**2)-(2*x)+2)
F3 = sp.Integral(f3,(x,1,sp.pi))
print(F3.evalf())
```

```python
# Problem 4
f4 = ((2*x)+1)/(sp.sqrt(x*(1-x)))
F4 = sp.Integral(f4,(x,0,1))
print(F4.evalf())
```

```python
# Problem 5
f5 = x*y*(2-(x**2))*(2-(x*y))
F51 = sp.integrate(f5,(x,(y-4)/2,(y+4)/2))
F52 = sp.integrate(F51,(y,-2,2))
print(F52.evalf())
```

```python
# Ben Kavouras
# A couple modules to get started
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# define a simple function
def f(x):
    f = 2*np.sin(0.5*np.pi*x)+3*np.cos(0.3*np.pi*x)
    return f
```

```python
# a simple 2D function
def f2d(x,y):
    f2d = 2*np.sin(0.5*np.pi*x)*np.cos(0.3*np.pi*x)+3*np.sin(0.5*np.pi*y)*np.cos(0.3*np.pi*y)
    return f2d
```

```python
# a simple line plot
x = np.linspace(0,10*np.pi, 100)
fig, ax = plt.subplots()
ax.plot(x, f(x), 'p')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('Wavy function, f(x)')
ax.set_xticks(np.arange(0,10*np.pi,np.pi))
```

```python
# a simple contour plot
i = 10
x = np.linspace(0,10*np.pi,100)
y = np.linspace(0,10*np.pi,100)
X,Y = np.meshgrid(x,y)

fig, ax = plt.subplots()
cp = ax.contour(X,Y,f2d(X,Y))
```
